```python
import pandas as pd

train_df = pd.read_csv("BASA_cricket_train.tsv", sep = "\t")
test_df = pd.read_csv("BASA_cricket_test.tsv", sep = "\t")
```

```python
def class_to_numeric(df_column):
  if df_column == "neutral":
    return 0
  elif df_column == "positive":
    return 1
  elif df_column == "negative":
    return 2
```

```python
train_labels = pd.Series(train_df["class_label"].apply(class_to_numeric))
```

```python
len(train_labels)
```

```python
train_df.loc[1]
```

```python
train_df["labels"] = train_labels
train_df.head()
```

```python
test_labels = pd.Series(test_df["class_label"].apply(class_to_numeric))
test_df["labels"] = test_labels
test_df.head()
```

```python
len(test_df) , len(train_df)
```

```python
data_files = {
    "train": "train_df",
    "test": "test_df"
}
```

```python
!pip install datasets --quiet
```

```python
from datasets import load_dataset

raw_datasets = load_dataset("pandas", data_files = data_files)
raw_datasets
```

```python
from datasets import Dataset
train_dataset = Dataset.from_pandas(train_df)
train_dataset
```

```python
test_dataset = Dataset.from_pandas(test_df)
test_dataset
```

```python
from transformers import AutoTokenizer
checkpoint = "monsoon-nlp/bangla-electra"

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
tokenizer
```

```python
def tokenize_function(example):
  return tokenizer(
      example["text"],
      truncation = True
  )
```

```python
tokenized_train_dataset = train_dataset.map(tokenize_function, batched = True)
tokenized_test_dataset = test_dataset.map(tokenize_function, batched = True)
```

```python
tokenized_train_dataset[0]
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments("test-trainer")
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(checkpoint,num_labels = 3)
model.config
```

```python
from transformers import Trainer
trainer = Trainer(
    model, 
    training_args,
    train_dataset = tokenized_train_dataset,
    eval_dataset = tokenized_test_dataset,
    tokenizer = tokenizer,
)
```

```python
trainer.train()
```

```python
predictions = trainer.predict(tokenized_test_dataset)
predictions
```
