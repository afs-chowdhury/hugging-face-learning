```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

classifier("I'm not a very bad boy !")
```

```python
classifier(["I love ohi.","I would love to marry her and live with her very soon."])
```

```python
from transformers import pipeline

z_s_classifier = pipeline("zero-shot-classification")

z_s_classifier(
    ["I am quite good at badminton.","I love to study"],
    candidate_labels = ["business","education","affection", "sports"]

)
```

```python
generator = pipeline("text-generation",model = "xlnet-base-cased")

generator("I am not hungry and i don't want",
          num_return_sequences = 2
          
)
```

```python
unmasker = pipeline("fill-mask")

unmasker("I <mask> you OHI with all my heart.",
         top_k = 3)
```

```python
b_b_c_unmasker = pipeline("fill-mask", model = "bert-base-cased")
b_b_c_unmasker("I [MASK] you OHI.",
               top_k = 2)
```

```python
ner = pipeline("ner", grouped_entities = True)
ner("My name is shagoto and I study information and commutination engineering at Islamic University, Kushtia")
```
