```python
## changing the config file and saving  it to another model

from  transformers import BertConfig, BertModel

config_1 = BertConfig.from_pretrained('ProsusAI/finbert',num_hidden_layers=8)
model_3 = BertModel(config_1)
model_2.config, model_3.config
```

```python
#saving the model
model_3.save_pretrained("my-bert-model")
```

```python
#loading the saved model

model_4 = BertModel.from_pretrained("my-bert-model")
model_3.config,model_4.config
```
