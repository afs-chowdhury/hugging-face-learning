```python
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

model_inputs = tokenizer(sequences)
model_inputs
```

## padding 


```python

# padding with maximum sequence length
model_inputs_1 = tokenizer(sequences,padding = "longest")
model_inputs_1
```

```python
# padding with maximum length of the model 

model_inputs_2 = tokenizer(sequences, padding  = "max_length")
model_inputs_2
```

```python
# padding with max length 

tokenizer(sequences, padding = "max_length", max_length = 8)
```

### Truncation


```python
# truncation with max length of the model
tokenizer(sequences, truncation = True)
```

```python
# truncation with defined max length 
tokenizer(sequences, truncation = True, max_length = 6)
```

```python
### return as tensor and numpy arrays
```

```python
# pytorch tensor
tokenizer(sequences, padding = True, truncation = True, return_tensors = "pt")
```

```python
# tensorflow tensors
tokenizer(sequences, padding = True, truncation = True, return_tensors = "tf")
```

```python
# numpy arrays 
tokenizer(sequences , padding = True, truncation = True, return_tensors = "np")
```
